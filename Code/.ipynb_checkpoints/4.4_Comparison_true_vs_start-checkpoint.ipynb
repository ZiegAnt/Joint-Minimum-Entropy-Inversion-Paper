```python
# Importing pygimli
import  pygimli                 as     pg
import  pygimli.meshtools       as     mt
from    pygimli.viewer.mpl      import createColorBarOnly

# Importing auxiliary libraries
import  numpy                   as     np
import  matplotlib.pyplot       as     plt
import  sys


# Plotting
c_ert = 'gnuplot'
c_srt = 'viridis'

#Mapping matrial properties to units
rhomap = [[1,700], [2,700], [3,70], [4,7000]]
vmap = [[1,500], [2,4000], [3,3000], [4,2000]]
```

## Geometry for Comparison

```python
# Create layered subsurface
world = mt.createWorld([-35,-20], [35, 0], layers=[-13],marker=1)
cube = mt.createRectangle(pos=[-8, -8], size=[9.0, 9.0],marker=3)
circ = mt.createCircle(pos=(8, -8), radius=4.5, nSegments=30,marker=4)
geom = world + circ + cube
geom_mesh = mt.createMesh(geom, area=0.5)

r_true = np.array(geom_mesh.cellMarkers())
r_true[r_true == 1] = 700
r_true[r_true == 2] = 700
r_true[r_true == 3] = 70
r_true[r_true == 4] = 7000

v_true = np.array(geom_mesh.cellMarkers())
v_true[v_true == 1] = 500
v_true[v_true == 2] = 4000
v_true[v_true == 3] = 3000
v_true[v_true == 4] = 2000

r_start = pg.Vector(geom_mesh.cellCount(),619)
v_start = 500-4000/20*np.array(geom_mesh.cellCenters())[:,1]

v_new = np.array(geom_mesh.cellMarkers())
v_new[v_new == 1] = 500
v_new[v_new == 2] = 4000
v_new[v_new == 3] = 500
v_new[v_new == 4] = 500
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,2))
fig.tight_layout(pad=0.5)

ax[0].set_title(r'$|\rho_{start}-\rho_{true}|$', fontsize = 14)
ax[1].set_title(r'$|v_{start}-v_{true}|$', fontsize = 14)

pg.show(geom_mesh, abs(v_true-v_start), ax=ax[1], cMap=c_srt, cMin=0, cMax=3000, colorBar=False, logScale=False)

pg.show(geom_mesh, abs(r_true-r_start), ax=ax[0], cMap=c_ert, cMin=0, cMax=3000, colorBar=False, logScale=False)


    # Adjust axis labels
ax[0].set_ylabel('Z (m)')
ax[0].set_xlabel('X (m)')
ax[1].set_xlabel('X (m)')
ax[1].set_yticks([])
for axis in ax:
        axis.set_xlim(-35,35)
        axis.set_ylim(-20,0)

    # Add colorbar ERT
cax1 = ax[0].inset_axes([0.05, -0.6, 0.9, 0.15])
createColorBarOnly(ax=cax1, cMin=0, cMax=3000, logScale=False,cMap=c_ert,
                      label=pg.unit('res'), orientation='horizontal')

    # Add colorbar SRT
cax2 = ax[1].inset_axes([0.05, -0.6, 0.9, 0.15])
createColorBarOnly(ax=cax2, cMin=0, cMax=3000, logScale=False,cMap=c_srt,
                      label=pg.unit('vel'), orientation='horizontal')

plt.savefig(f'C:/Users/azieg/Desktop/GIT_repositories/APG-MSc-Project-Ziegon/Figures/Ch-6/Starting_true_contrast.pdf', 
             bbox_inches='tight',
             dpi=600)
```
